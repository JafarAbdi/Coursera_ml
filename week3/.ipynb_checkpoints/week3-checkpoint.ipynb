```python
def Gradient(w, X, Y):
    w = w.reshape(w.shape[0], 1)
    m = X.shape[0]
    z  = np.dot(X, w).reshape(X.shape[0], 1)
    dJ = (1 / m) * np.sum((sigmoid(z) - Y) * X, axis=0)
    return dJ

def CostFunction(w, X, Y):
    w = w.reshape(w.shape[0], 1)
    m = X.shape[0]
    z = np.dot(X, w)
    J = (-1 / m) * np.sum((Y * np.log(sigmoid(z)) + (1 - Y) * np.log(1 - sigmoid(z))))
    return J
```

```python
theta, c = scipy.optimize.fmin_bfgs(f=CostFunction, x0=[0.,0.,0.], fprime=Gradient, args=(X, Y), retall=1)
```

```python
# degree = 6
# OUT = np.ones_like(X[:, 0]).reshape(X.shape[0], 1)
# for i in range(1, degree + 1):
#     for j in range(i + 1):
#         OUT = np.concatenate((OUT, ((X[:,1]**(i - j)) * (X[:,2]**j)).reshape(X.shape[0], 1)),axis=1)
def mapFeature(X1, X2, degree):
    m   = X1.shape[0]
    X1  = X1.reshape(m, 1)
    X2  = X2.reshape(m, 1)
    OUT = np.ones((m, 1))
    for i in range(1, degree + 1):
        for j in range(i + 1):
            OUT = np.concatenate((OUT, ((X1**(i - j)) * (X2**j))),axis=1)
    return OUT
```

```python
w = np.zeros((OUT.shape[1],))
```

```python
def Gradient(w, X, Y, lam):
    w = w.reshape(w.shape[0], 1)
    m = X.shape[0]
    z  = np.dot(X, w).reshape(X.shape[0], 1)
    dJ = (1 / m) * np.sum((sigmoid(z) - Y) * X, axis=0) + (lam / m) * np.hstack((0, w[1:, 0]))
    return dJ

def CostFunction(w, X, Y, lam):
    w = w.reshape(w.shape[0], 1)
    m = X.shape[0]
    z = np.dot(X, w)
    J = (-1 / m) * np.sum((Y * np.log(sigmoid(z)) + (1 - Y) * np.log(1 - sigmoid(z)))) + (lam / (2 * m)) * np.dot(w[1:, 0], w[1:, 0])
    return J
```
